```python
from random import randint
from sklearn.linear_model import LinearRegression
from tabulate import tabulate
import numpy as np
import time
start_time = time.time()

TRAIN_SET_LIMIT = 10
TRAIN_SET_COUNT = 200
TRAIN_INPUT = list()
TRAIN_OUTPUT = list()
tdata=list()   
asum=int()
c=int()
op_av=int()
for i in range(TRAIN_SET_COUNT):
    a = randint(0,TRAIN_SET_LIMIT)
    b = randint(0,TRAIN_SET_LIMIT)
    c = randint(0,TRAIN_SET_LIMIT)
    op = (2*a) + b - c
    #op=randint(0,TRAIN_SET_LIMIT)
    TRAIN_INPUT.append([a, b, c])
    TRAIN_OUTPUT.append(op)
    tdata.append([a,b,c,op]) 


predictor = LinearRegression(n_jobs=-1)
predictor.fit(X=TRAIN_INPUT, y=TRAIN_OUTPUT)
X_TEST = [[5, 9, 10]]
outcome = predictor.predict(X=X_TEST)
coefficients = predictor.coef_
c=0
for i in range(TRAIN_SET_COUNT):
    dataset=tdata[i]
    if dataset[0]==5 and dataset[1]==9 and dataset[2]==10:
        print (dataset)
        asum=asum + dataset[3]
        c=c+1
    
if c!=0:
    op_av=asum/c
print ("average of the outputs",op_av)
print ("total matching data",c)

print(tabulate(tdata,headers=["a","b","c","output"]))
np.savetxt("trainingdata.csv",tdata,delimiter="\t", fmt='%s',header="#a\tb\tc\toutput",comments='')
print('Outcome : {}\nCoefficients : {}'.format(outcome, coefficients))

print("--- %s seconds ---" % (time.time() - start_time))
```
