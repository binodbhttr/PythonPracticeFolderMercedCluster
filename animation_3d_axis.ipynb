```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation


raw = np.random.rand(100,3)
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
x = raw[:, 0]
y = raw[:, 1]
z = raw[:, 2]

ax.scatter(x, y, -z, zdir='z', c='black', depthshade=False, s=1, marker='.')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

def rotate(angle): 
    '''angle would automatically take each value of frame
    elev when take values (0-360) rotates in the direction of right hand rule thumb pointing left of the screen
    azim when takes values (0-360) roates in the direction of right hand rule thumb pointing down
    set elev to 0 if you don't want rotation in that direction 
    if you set any constant number to elev or azim, it would remain the same in each frame
    '''
    print(angle)
    ax.view_init(elev=angle,azim=0) #

print("Making animation")
rot_animation = animation.FuncAnimation(fig, rotate, frames=np.arange(0, 362, 2), interval=100)
rot_animation.save('rotation.gif', dpi=80, writer='imagemagick')
```
