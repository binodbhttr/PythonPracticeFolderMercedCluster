```python
import numpy as np
import scipy as sp

from matplotlib import pyplot as plt
from random import randint
n=100
x=[i for i in range(1,n)]

y=[randint(-1000,1000) for i in range(1,n)]
    
#plt.hist(y,bins=5)

plt.plot(x,y)
plt.ylabel("Random Numbers between -1000 to 1000")
plt.xlabel("Time")
plt.show()
plt.close()
```
