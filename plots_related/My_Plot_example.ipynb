```python
from matplotlib import pyplot as plt

#create a figure object for each figure you want:

fig1=plt.figure()

#then add axes to the figure. You can have as many axes as you want
ax1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)

#here we created a 2 by 2 (4 subplots inside a figure)
# you can set xlabel and ylabel as
ax1.set_xlabel("This is the xlabel for the first axes")
ax2.set_ylabel("This is the ylabel for the second axes")
ax3.set_title("This is the title for the third axes")
#note the labels may not appear properly in the subplots
# use fig1.tight_layout() before saving or showing the figure
fig1.tight_layout()
#finally to save the figure
fig1.savefig("filename.png")
#you can also save in any other formats you want.

# Now to create a new figure
fig2=plt.figure()
x1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)
fig2.tight_layout()
#similarly add axes and continue
```
