```python
from matplotlib import pyplot as plt

# Subplots are organized in a Rows x Cols Grid
# Tot and Cols are known

tot = 20
cols = 3

# Compute Rows required

rows = tot // cols #Integer divide total by column Not a problem if perfectly divisible
rows += tot % cols

# Create a Position index
x=[1,2,3,4,5]
y=[5,7,8,9,2]
position = range(1,tot + 1)
# Create main figure

fig = plt.figure(figsize=(15,30))
for k in range(tot): # add every single subplot to the figure with a for loop
    ax = fig.add_subplot(rows,cols,position[k])
    ax.scatter(x,y,marker=".",s=10)
    ax.plot(5,9,color='red',marker=".")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_xlim(3,4)
    ax.set_ylim(4,7)
    title="Snapshot " + str(k)
    ax.set_title(title)
plt.tight_layout()
plt.show()
```
